# Criando Coluna de Estação para o df e csv

```python
queimadas_por_bioma_mes['Ano_Mes'] = pd.to_datetime(queimadas_por_bioma_mes['Ano_Mes'])

# Extraindo Ano e Mês
queimadas_por_bioma_mes['Ano'] = queimadas_por_bioma_mes['Ano_Mes'].dt.year
queimadas_por_bioma_mes['Mes'] = queimadas_por_bioma_mes['Ano_Mes'].dt.month


meses_para_estacoes = {
    12: 'Verão', 1: 'Verão', 2: 'Verão',
    3: 'Outono', 4: 'Outono', 5: 'Outono',
    6: 'Inverno', 7: 'Inverno', 8: 'Inverno',
    9: 'Primavera', 10: 'Primavera', 11: 'Primavera'
}


queimadas_por_bioma_mes['Estacao'] = queimadas_por_bioma_mes['Mes'].map(meses_para_estacoes)


focos_por_mes = queimadas_por_bioma_mes.groupby(['Ano', 'Mes', 'Estacao'])['Focos'].sum().reset_index()
focos_por_mes = focos_por_mes.sort_values(by=['Ano', 'Mes'], ascending=[True, True])


print(focos_por_mes)
```

```python
# Salvar o DataFrame como CSV
focos_por_mes.to_csv('queimadas_atual_2.0.csv', index=False)
```

```python
# Gráfico de Barras com paleta de cores para todas as combinações de Ano, Estação e Bioma
#grupar os dados por Ano, Mês e Estação para somar os focos de queimadas
focos_por_mes = queimadas_por_bioma_mes.groupby(['Ano', 'Mes', 'Estacao'])['Focos'].sum().reset_index()

# Criar uma nova coluna combinando Ano e Estação para o eixo de agrupamento
focos_por_mes['Ano_Estacao'] = focos_por_mes['Ano'].astype(str) + ' - ' + focos_por_mes['Estacao']

# Definindo a paleta dinamicamente
unique_comb = focos_por_mes['Ano_Estacao'].unique()
palette_estacoes = sns.color_palette("husl", len(unique_comb))

# Criar um dicionário para mapear as combinações para as cores
color_mapping = dict(zip(unique_comb, palette_estacoes))

# Configurar o gráfico
plt.figure(figsize=(14, 7))
sns.barplot(data=focos_por_mes, x='Mes', y='Focos', hue='Ano_Estacao', palette=color_mapping, ci=None)
plt.title('Focos de Queimadas por Mês e Estação do Ano (Comparação entre 2022 e 2023)', fontsize=16, fontweight='bold')
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Número de Focos', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Ano - Estação', fontsize=10)
plt.tight_layout()
plt.show()



```

# 3º passo: Análise Estatística -Teste ANOVA

Para investigar a relação entre a frequência de queimadas e as estações do ano, utilizamos o teste ANOVA (Análise de Variância). <br>O objetivo foi determinar se há diferenças significativas nas médias de focos de queimadas entre as diferentes estações do ano (verão, outono, inverno e primavera). <br>Assim, além de analisar se certos biomas são mais propensos a queimadas, também buscamos entender se as variações sazonais influenciam essa ocorrência.

```python
from scipy import stats

grupos = [grupo['Focos'].values for nome, grupo in focos_por_mes.groupby('Estacao')]

# Realizar ANOVA
anova_result = stats.f_oneway(*grupos)

print("Resultados da ANOVA:")
print(f"Estatística F: {anova_result.statistic}")
print(f"Valor p: {anova_result.pvalue}")

# Conclusão
alpha = 0.05
anova_conclusao = ""
if anova_result.pvalue < alpha:
    anova_conclusao = "Rejeitamos a hipótese nula: há diferenças significativas entre as estações."
else:
    anova_conclusao = "Não rejeitamos a hipótese nula: não há diferenças significativas entre as estações."

```

#1º Passo: Categorizar os perídos de seca

```python
# Converter a coluna DataHora para o tipo datetime
df_agrupado['DataHora'] = pd.to_datetime(df_agrupado['DataHora'])

# Criar uma coluna de mês/ano
df_agrupado['MesAno'] = df_agrupado['DataHora'].dt.to_period('M')

# Criar uma nova coluna para categorizar os períodos de seca
df_agrupado['CategoriaSeca'] = pd.cut(df_agrupado['DiaSemChuva'], bins=[-1, 5, 10, 20, df_agrupado['DiaSemChuva'].max()],
                             labels=['Curta (<=5 dias)', 'Moderada (6-10 dias)', 'Longa (11-20 dias)', 'Extensa (>20 dias)'])

# Contar a quantidade de focos de queimadas para cada categoria de seca por mês/ano
contagem_queimadas_por_categoria_mesano = df_agrupado.groupby(['MesAno', 'CategoriaSeca'])['foco_id'].count().unstack()

# Mostrar o resultado
contagem_queimadas_por_categoria_mesano.head(25)
```

#2º Passo: Visualização dos focos de queimadas por categoria de seca e ano

```python

# Separar os dados por ano e categoria de seca
df_agrupado['Ano'] = pd.to_datetime(df_agrupado['DataHora']).dt.year

# Agrupar por ano e categoria de seca para contar focos de queimadas
contagem_queimadas_por_categoria_mesano = df_agrupado.groupby(['Ano', 'CategoriaSeca'])['foco_id'].count().reset_index()

# Gráfico de barras para comparar a quantidade de queimadas entre os anos e categorias
plt.figure(figsize=(10, 6))
sns.barplot(x='CategoriaSeca', y='foco_id', hue='Ano', data=contagem_queimadas_por_categoria_mesano, palette='viridis')

# Ajustar o título e os rótulos dos eixos
plt.title('Quantidade de Focos de Queimadas por Categoria de Seca e Ano', fontsize=14)
plt.xlabel('Categoria de Seca', fontsize=12)
plt.ylabel('Quantidade de Queimadas', fontsize=12)
plt.xticks(rotation=0)


# Exibir o gráfico
plt.xticks(rotation=0, fontsize=8)
plt.yticks(fontsize=8)
plt.show()
```

#3º Passo: Visualização dos focos de queimadas por categoria de seca e ano, nos diferentes biomas.

De acordo com o INPE: "No América do Sul, a quase totalidade das queimadas é causada pelo Homem, por razões muito variadas: limpeza de pastos, preparo de plantios, desmatamentos, colheita manual de cana-de-açúcar, vandalismo, balões de São João, disputas fundiárias, protestos sociais, e etc. Com mais de 300.000 queimadas e nuvens de fumaça cobrindo milhões de km2 detectadas anualmente através de satélites, o País ocupa o 5º lugar entre os países poluidores, e também devastando anualmente em média cerca de 15 mil km2/ano de florestas naturais.

Queimadas naturais são aquelas causadas por raios, e portanto inexistentes nos períodos de estiagem prolongada e intensa, que é quando mais se queima no País. Elas são mais comuns no bioma Cerrado, que é adaptado ao fogo, e ocorrem em geral no final e início da estação chuvosa, afetando as mesmas áreas a cada oito a dez anos.

José Bonifácio de Andrada e Silva, em ~1820, resumiu as razões para as nossas queimadas e desmatamentos: "ignorância, associada à preguiça e má fé". Nos dias de hoje, em que todos sabem dos efeitos negativos das queimadas, temos apenas de inverter a ordem dos motivos para: má fé, associada à preguiça e ignorância."

```python
# Converter a coluna DataHora para o tipo datetime
df_agrupado['DataHora'] = pd.to_datetime(df_agrupado['DataHora'])

# Criar uma coluna de mês/ano
df_agrupado['MesAno'] = df_agrupado['DataHora'].dt.to_period('M')

# Criar uma nova coluna para categorizar os períodos de seca
df_agrupado['CategoriaSeca'] = pd.cut(df_agrupado['DiaSemChuva'], bins=[-1, 5, 10, 20, df_agrupado['DiaSemChuva'].max()],
                             labels=['Curta (<=5 dias)', 'Moderada (6-10 dias)', 'Longa (11-20 dias)', 'Extensa (>20 dias)'])

# Contar a quantidade de focos de queimadas para cada categoria de seca por mês/ano
contagem_queimadas_por_categoria_mesano = df_agrupado.groupby(['MesAno', 'CategoriaSeca'])['foco_id'].count().unstack()

# Mostrar o resultado
contagem_queimadas_por_categoria_mesano.head(10)  # Mostrar as primeiras 10 linhas para visualização
```

#4º Passo: Análise estatística: Anova oneway entre os períodos de seca de cada ano.

```python

# Dividir os dados em dois DataFrames: um para 2022 e outro para 2023
df_2022 = contagem_queimadas_por_categoria_mesano.loc[contagem_queimadas_por_categoria_mesano.index.year == 2022]
df_2023 = contagem_queimadas_por_categoria_mesano.loc[contagem_queimadas_por_categoria_mesano.index.year == 2023]

# Aplicar o teste ANOVA para 2022
anova_2022 = stats.f_oneway(
    df_2022['Curta (<=5 dias)'].dropna(),
    df_2022['Moderada (6-10 dias)'].dropna(),
    df_2022['Longa (11-20 dias)'].dropna(),
    df_2022['Extensa (>20 dias)'].dropna()
)

# Aplicar o teste ANOVA para 2023
anova_2023 = stats.f_oneway(
    df_2023['Curta (<=5 dias)'].dropna(),
    df_2023['Moderada (6-10 dias)'].dropna(),
    df_2023['Longa (11-20 dias)'].dropna(),
    df_2023['Extensa (>20 dias)'].dropna()
)

# Mostrar os resultados
print("ANOVA para 2022")
print("Estatística F:", anova_2022.statistic)
print("Valor p:", anova_2022.pvalue)

print("\nANOVA para 2023")
print("Estatística F:", anova_2023.statistic)
print("Valor p:", anova_2023.pvalue)
```

#Conclusão da da hipotese 2:

h0: A ocorrência de queimadas não tem influência do tamanho dos períodos de seca/sem precipitação.

A análise de variância para o ano de 2022 não apresentou diferença significativa (p>0,05) entre os focos de queimadas e os diferentes períodos de seca.

Enquanto para o ano de 2023 a análise de variância apresentou diferença significativa (p<0,05) entre os focos de queimadas e os diferentes períodos de seca.

Não houve tempo hábil para realizar a pesquisa para aplicar um teste a posteriori.

#1º Passo: Agrupar os dados dos focos das queimadas e da precipitação média diária mensal, por ano.

```python
# Agrupar os dados por mês/ano
df_agrupado['Mes'] = df_agrupado['DataHora'].dt.month

# Agrupar os dados por mês/ano para focos de queimadas
focos_mensal = df_agrupado.groupby(['Ano', 'Mes'])['foco_id'].count().reset_index()

# Agrupar os dados para calcular a precipitação média por mês/ano
precipitacao_mensal = df_agrupado.groupby(['Ano', 'Mes'])['Precipitacao'].mean().reset_index()
```

2º Passo: Visualização dos focos de queimadas mensais, por ano. E visualização da precipítação média diária a cada mês, por ano.

```python
# Definir o tamanho da figura
plt.figure(figsize=(12, 6))

# Configurar o eixo principal com barras (focos de queimadas)
# Aqui, utilizamos tons de vermelho e marrom para os anos 2022 e 2023
ax1 = sns.barplot(
    data=focos_mensal, x='Mes', y='foco_id', hue='Ano',
    palette={2022: '#8B0000', 2023: '#A52A2A'}, alpha=0.7
)
ax1.set_xlabel('Mês')
ax1.set_ylabel('Quantidade de Focos de Queimadas')
ax1.set_title('Focos de Queimadas e Precipitação Média Mensal (2022 e 2023)')

# Configurar o segundo eixo para as linhas (precipitação)
# Aqui, utilizamos tons de azul claro e azul escuro para os anos 2022 e 2023
ax2 = ax1.twinx()
ax2 = sns.lineplot(
    data=precipitacao_mensal, x='Mes', y='Precipitacao', hue='Ano',
    marker='o', ax=ax2, palette={2022: '#1E90FF', 2023: '#00008B'}
)
ax2.set_ylabel('Precipitação Média (mm)')

# Ajustar as legendas e evitar sobreposição
ax1.legend(loc='upper left', title='Ano - Focos de Queimadas')
ax2.legend(loc='upper right', title='Ano - Precipitação')

plt.tight_layout()
plt.show()
```

#2º Passo: Análise estatística: Correlção de Pearson

União dos df de foco de queimadas e precipitação pelas colunas mês e ano, para termos a precipitação média e o número de focos de queimadas por mês no mesmo DataFrame.

Calculo da correlação de Pearson entre os focos de queimadas e a precipitação média diária de cada mês, por ano.


```python
#combinar os dados de focos de queimadas e precipitação
dados_comb = pd.merge(focos_mensal, precipitacao_mensal, on=['Mes', 'Ano'])

#separar os dados por ano
dados_2022 = dados_comb[dados_comb['Ano'] == 2022]
dados_2023 = dados_comb[dados_comb['Ano'] == 2023]

# Calcular a correlação de Pearson para 2022
correlacao_2022 = dados_2022['foco_id'].corr(dados_2022['Precipitacao'])

# Calcular a correlação de Pearson para 2023
correlacao_2023 = dados_2023['foco_id'].corr(dados_2023['Precipitacao'])

# Exibir os resultados
print(f"Correlação entre Focos de Queimadas e Precipitação em 2022: coeficiente de correlação de Pearson r={correlacao_2022:.4f}")
print(f"Correlação entre Focos de Queimadas e Precipitação em 2023: coeficiente de correlação de Pearson r={correlacao_2023:.4f}")
```
