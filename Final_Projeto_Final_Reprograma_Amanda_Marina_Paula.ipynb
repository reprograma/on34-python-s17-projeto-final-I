```python
from matplotlib import pyplot as plt
import seaborn as sns
saude_tec_port.groupby('Empresa_Tecnologia').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#Trabalhadores em empresas de tecnologia e que possuem "Sim ou Talvez" em "Consequencia_Saude_Mental"  X
# Trabalhadores que NÃO trabalham em empresas de tecnologia  e que possuem "Sim ou Talvez" em "Consequencia_Saude_Mental"

trabalhadores_tech = saude_tec_port[
    (saude_tec_port['Empresa_Tecnologia'] == 'Sim')
]

trabalhadores = saude_tec_port[
    (saude_tec_port['Empresa_Tecnologia'] == 'Não')
]

trabalhadores_depre = trabalhadores[
    (trabalhadores['Consequencia_Saude_Mental'] == 'Sim') | (trabalhadores['Consequencia_Saude_Mental'] == 'Talvez')
]
trabalhadores_depre_count = len(trabalhadores_depre) / len(trabalhadores)
trabalhadores_tech_depre_count = len(trabalhadores_depre) / len(trabalhadores_tech)

categorias = ['Trabalham em Empresas de Tecnologia', 'Não Trabalham em Empresas de Tecnologia']
medias = [trabalhadores_tech_depre_count, trabalhadores_depre_count]

plt.figure(figsize=(8, 6))
plt.bar(categorias, medias, color=['blue', 'orange'])
plt.title('Média de Consequências na Saúde Mental')
plt.ylabel('Proporção de Pessoas com "Sim" ou "Talvez')

# Exibir o gráfico
plt.show()
```

```python
#O trabalho remoto é melhor para a saúde mental?

trabalhadores_depre = trabalhadores[
    (trabalhadores['Consequencia_Saude_Mental'] == 'Sim') | (trabalhadores['Consequencia_Saude_Mental'] == 'Talvez')
]

trabalhadores_no_depre = trabalhadores[
    (trabalhadores['Consequencia_Saude_Mental'] == 'Não')
]
trabalhadores_remoto = trabalhadores[
    (trabalhadores['Trabalho_Remoto'] == 'Sim')
]


trabalhadores_remoto_depre = len(trabalhadores_remoto)/len(trabalhadores_depre)
trabalhadores_remoto_no_depre = len(trabalhadores_remoto)/len(trabalhadores_no_depre)

categorias = ['Trabalhadores remoto', 'Trabalhadores presenciais']
medias = [trabalhadores_remoto_depre, trabalhadores_remoto_no_depre]

plt.figure(figsize=(8, 6))
plt.bar(categorias, medias, color=['blue', 'orange'])
plt.title('Média de Consequências na Saúde Mental')
plt.ylabel('Média de Trabalhadores')

# Exibir o gráfico
plt.show()
```

```python
#Filtrar MULHERES que trabalham no setor tec "Empresa_Tecnologia"
mulheres_tec = uniao_bases[(uniao_bases['Genero'] == 'Feminino') & (uniao_bases['Empresa_Tecnologia'] == 'Sim')]

plt.figure(figsize=(10, 6))
sns.countplot(data=mulheres_tec, x='Genero', hue='Consequencia_Saude_Mental')
plt.title('Prevalência de Transtornos Mentais por Gênero Fora do Setor de Tecnologia')
plt.show()
```

```python
#Filtrar apenas quem trabalha no setor de tecnologia  "Empresa_Tecnologia"
trabalha_tec = uniao_bases[uniao_bases['Empresa_Tecnologia'] == 'Sim']


plt.figure(figsize=(10, 6))
sns.countplot(data=trabalha_tec, x='Genero', hue='Consequencia_Saude_Mental')
plt.title('Prevalência de Transtornos Mentais por Gênero no Setor de Tecnologia')
plt.show()

```

```python
# Trocar campos "Sim, talvez e possivelmente" para que ele se torne um número
uniao_bases['Consequencia_Saude_Mental'] = uniao_bases['Consequencia_Saude_Mental'].replace({
    'Sim': 1,
    'Talvez': 1,
    'Possivelmente': 1,
    'Não sei': 0,
    'Não': 0
})

# Filtrar homens e mulheres que trabalham no setor tec
trabalha_tec = uniao_bases[((uniao_bases['Genero'] == 'Masculino') | (uniao_bases['Genero'] == 'Feminino')) & (uniao_bases['Empresa_Tecnologia'] == 'Sim')]

# Filtrar homens e mulheres que NÃO trabalham no setor tec
nao_trabalha_tec = uniao_bases[((uniao_bases['Genero'] == 'Masculino') | (uniao_bases['Genero'] == 'Feminino')) & (uniao_bases['Empresa_Tecnologia'] == 'Não')]

# Calcular a média de "Consequência para a Saúde Mental" (1 para Sim, 0 para Não) para cada grupo
trabalha_tec_media = trabalha_tec.groupby('Genero')['Consequencia_Saude_Mental'].mean()
nao_trabalha_tec_media = nao_trabalha_tec.groupby('Genero')['Consequencia_Saude_Mental'].mean()

# Exibir as médias
print("Média de Consequência para a Saúde Mental para quem trabalha no setor tec:")
print(trabalha_tec_media)

print("\nMédia de Consequência para a Saúde Mental para quem NÃO trabalha no setor tec:")
print(nao_trabalha_tec_media)

```

```python
# Calcular médias
trabalha_tec_media = trabalha_tec.groupby('Genero')['Consequencia_Saude_Mental'].mean().reset_index()
trabalha_tec_media['Setor'] = 'Trabalha no setor Tec'

nao_trabalha_tec_media = nao_trabalha_tec.groupby('Genero')['Consequencia_Saude_Mental'].mean().reset_index()
nao_trabalha_tec_media['Setor'] = 'Não trabalha no setor Tec'

# Combinar os campos
medias_setor_tec = pd.concat([trabalha_tec_media, nao_trabalha_tec_media])

# Criar gráfico
plt.figure(figsize=(7, 7))
sns.barplot(x='Genero', y='Consequencia_Saude_Mental', hue='Setor', data=medias_setor_tec)
plt.title('Média de Consequência para a Saúde Mental por Gênero e Setor de Trabalho')
plt.ylabel('Média de Consequência para a Saúde Mental')
plt.xlabel('Gênero')
plt.legend(title='Setor de Trabalho')
plt.show()
```
